```python
df = data[['RXDRSC1', 'BPQ080', 'PAQ650', 'HSD010', 'INQ030', 'INQ080']]
df = df.dropna()

target_column = 'RXDRSC1'


df.shape
```

```python
target = df[target_column]
data_ML = df.loc[:, df.columns != target_column]
```

```python
from sklearn.model_selection import train_test_split

#split data set into train and test sets
data_train, data_test, target_train, target_test = train_test_split(data_ML,target, test_size = 0.30, random_state = 10)

```

```python
from sklearn.linear_model import Ridge, RidgeCV, Lasso, LassoCV
from sklearn.metrics import mean_squared_error
from sklearn.metrics import accuracy_score

# lassocv = LassoCV(alphas = None, cv = 10, max_iter = 100000, normalize = True)
# lassocv.fit(data_train, target_train)

# lasso = Lasso(max_iter = 10000, normalize = True)
# lasso.set_params(alpha=lassocv.alpha_)
# lasso.fit(data_train, target_train)
# mean_squared_error(target_test, lasso.predict(data_test))

# pred = lasso.fit(data_train, target_train).predict(data_test)

lasso = LassoCV().fit(data_train, target_train)
pred = lasso.predict(data_test)
accuracy_score(target_test, pred.round(), normalize = True)
# mean_squared_error(target_test, pred)

```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

pred = DecisionTreeClassifier().fit(data_train, target_train).predict(data_test)
# print('Accuracy of Decision Tree classifier on training set: {:.2f}'
#      .format(clf.score(data_train, target_train)))
# print('Accuracy of Decision Tree classifier on test set: {:.2f}'
#      .format(clf.score(data_test, target_test)))

accuracy_score(target_test, pred, normalize = True)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score

gaussian_nb = GaussianNB()
pred = gaussian_nb.fit(data_train, target_train).predict(data_test)

print("Naive-Bayes accuracy : ", accuracy_score(target_test, pred, normalize = True))
```

```python
from sklearn.svm import LinearSVC
from sklearn.metrics import accuracy_score

linear_svc = LinearSVC(random_state=0)
pred = linear_svc.fit(data_train, target_train).predict(data_test)

print("LinearSVC accuracy : ", accuracy_score(target_test, pred, normalize = True))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

kneighbors = KNeighborsClassifier(n_neighbors=3)
kneighbors.fit(data_train, target_train)
pred = kneighbors.predict(data_test)

print ("KNeighbors accuracy score : ", accuracy_score(target_test, pred))
```
