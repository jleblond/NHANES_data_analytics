```python
import pandas as pd
import numpy as np


data = pd.read_csv('./data/cleaned/project_data.csv', index_col='SEQN')

data.head()
```

```python
# Removing rows where RIDAGEYR (age of person is NaN or age 0)

data = data[data.RIDAGEYR.notnull()]
# data = data[data.RIDAGEYR != 0]

data.shape
```

```python
data[['BPQ020']] = data[['BPQ020']].fillna(value=2)
data.head()
```

```python
corr = data.corr()
corr
```
